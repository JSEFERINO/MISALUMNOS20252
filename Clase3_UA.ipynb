```python
# prompt: Crea una tabla de frecuencias completa y el histograma con su poligono de frecuencia con la variable PESO

# @title **Paso 10. Tabla de frecuencias y histograma para la variable PESO**

# Calculate frequency table for 'PESO'
df['PESO'] = pd.to_numeric(df['PESO'], errors='coerce')
df_peso = df.dropna(subset=['PESO'])

# Calculate the number of bins (using Sturges' rule)
n = len(df_peso)
k = int(np.ceil(1 + 3.322 * np.log10(n)))

# Create frequency table
freq_table_peso = pd.cut(df_peso['PESO'], bins=k).value_counts().sort_index()
freq_table_peso = pd.DataFrame(freq_table_peso).reset_index()
freq_table_peso.columns = ['Intervalo de Peso', 'Frecuencia Absoluta']

# Calculate relative frequency
freq_table_peso['Frecuencia Relativa'] = freq_table_peso['Frecuencia Absoluta'] / n

# Calculate cumulative frequencies
freq_table_peso['Frecuencia Absoluta Acumulada'] = freq_table_peso['Frecuencia Absoluta'].cumsum()
freq_table_peso['Frecuencia Relativa Acumulada'] = freq_table_peso['Frecuencia Relativa'].cumsum()

print("Tabla de Frecuencias Completa para PESO:")
print(freq_table_peso)

# Create histogram
plt.figure(figsize=(10, 6))
sns.histplot(df_peso['PESO'], bins=k, kde=False, color='skyblue', edgecolor='black')

# Calculate bin centers for the frequency polygon
hist, bins = np.histogram(df_peso['PESO'], bins=k)
bin_centers = (bins[:-1] + bins[1:]) / 2

# Add frequency polygon
plt.plot(bin_centers, hist, marker='o', linestyle='-', color='red', label='Polígono de Frecuencia')

plt.title('Histograma y Polígono de Frecuencia de la variable PESO')
plt.xlabel('PESO')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# prompt: Construye el diagrama de caja de PARCIAL 2 vs SEXO vs ESTRATO

# Convert 'PARCIAL 2' to numeric, handling potential errors
df['PARCIAL 2'] = pd.to_numeric(df['PARCIAL 2'], errors='coerce')

# Drop rows where 'PARCIAL 2' is NaN after conversion
df_cleaned = df.dropna(subset=['PARCIAL 2'])

# Convert the 'ESTRATO' column to string type before getting unique values and sorting
# This handles potential float (NaN) values and ensures consistent sorting.
df_cleaned['ESTRATO'] = df_cleaned['ESTRATO'].astype(str)

# Sort 'ESTRATO' to ensure the order in the plot
estrato_order = sorted(df_cleaned['ESTRATO'].unique())

# Create the box plot
plt.figure(figsize=(12, 8))
sns.boxplot(x='ESTRATO', y='PARCIAL 2', hue='SEXO', data=df_cleaned, order=estrato_order)

plt.title('Diagrama de Caja de PARCIAL 2 por SEXO y ESTRATO')
plt.xlabel('ESTRATO')
plt.ylabel('PARCIAL 2')
plt.show()
```
